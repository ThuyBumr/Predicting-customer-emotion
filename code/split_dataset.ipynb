```python
from typing_extensions import Annotated
from os import path

root_path = '/content/gdrive/Shareddrives/BA/R Python/Đồ án/src'
dataset_path = path.join(root_path, 'dataset')
raw_path = path.join(dataset_path, 'raw')
annotated = path.join(dataset_path, 'annotated')

!ls "$dataset_path"
!ls "$annotated"
!ls "$raw_path"
```

```python
import pickle as pkl

with open(path.join(raw_path, 'electric.p'), 'rb') as f:
  electric_data = pkl.load(f)
```

```python
import pandas as pd

# Split the DataFrame into chunks of 1000 rows each
chunks = [electric_data[i:i+1000] for i in range(0, len(electric_data), 1000)]

# Save each chunk as an Excel file
for i, chunk in enumerate(chunks):
    chunk_df = pd.DataFrame(chunk)
    chunk_df.to_excel(path.join(annotated, f'eletric_{i}.xlsx'), index=False)
```

```python
import pandas as pd

df = pd.read_excel(path.join(annotated, "annotated_data.xlsx"))

rows_per_file = 2217

n_chunks = len(df) // rows_per_file

for i in range(n_chunks):
    start = i*rows_per_file
    stop = (i+1) * rows_per_file
    sub_df = df.iloc[start:stop]
    sub_df.to_excel(f"{annotated}/annotated-{i}.xlsx")
if stop < len(df):
    sub_df = df.iloc[stop:]
    sub_df.to_excel(f"{annotated}/annotatd-{i}.xlsx")
```
